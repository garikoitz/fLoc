```python
import os
import shutil
import random

# Define paths for the adult and child folders and the destination folder
adult_folder = '/media/tlei/data/toolboxes/fLoc/stimuli/body'  # Replace with actual path
child_folder =  '/media/tlei/data/toolboxes/fLoc/stimuli/limb'   # Replace with actual path
destination_folder = '/media/tlei/data/toolboxes/fLoc/stimuli/bodylimb'  # Replace with actual path

# Ensure the destination folder exists, if not, create it
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Get list of all files from the adult and child folders
adult_files = [os.path.join(adult_folder, file) for file in os.listdir(adult_folder) if file.endswith('.jpg')]
child_files = [os.path.join(child_folder, file) for file in os.listdir(child_folder) if file.endswith('.jpg')]

# Combine the lists of files
all_files = adult_files + child_files

# Shuffle the combined list of files randomly
random.shuffle(all_files)

# Move and rename the files to the destination folder as face-1.jpg, face-2.jpg, ..., face-288.jpg
for i, file_path in enumerate(all_files, start=1):
    new_file_name = f'bodylimb-{i}.jpg'
    new_file_path = os.path.join(destination_folder, new_file_name)
    
    # Move and rename the file
    shutil.copy(file_path, new_file_path)

    print(f"Copied and renamed: {file_path} -> {new_file_path}")

print("All files have been moved and renamed.")
```
